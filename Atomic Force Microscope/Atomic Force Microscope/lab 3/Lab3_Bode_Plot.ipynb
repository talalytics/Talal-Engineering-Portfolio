```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math

MagnitudePlant= pd.read_csv("/content/drive/MyDrive/advanced_mech2/lab 3/Lab_3.3_Plant_Amplitude.csv")
PhasePlant = pd.read_csv("/content/drive/MyDrive/advanced_mech2/lab 3/Lab_3.3_Plant_Phase.csv")

MagnitudeLoop= pd.read_csv("/content/drive/MyDrive/advanced_mech2/lab 3/Lab_3.4_Loop_Amplitude.csv")
PhaseLoop = pd.read_csv("/content/drive/MyDrive/advanced_mech2/lab 3/Lab_3.4_Loop_Phase.csv")

MagnitudeCL = pd.read_csv("/content/drive/MyDrive/advanced_mech2/lab 3/Lab_3.4_CL_Amplitude.csv")
PhaseCL = pd.read_csv("/content/drive/MyDrive/advanced_mech2/lab 3/Lab_3.4_CL_Phase.csv")

list(MagnitudePlant)
list(PhasePlant)
list(MagnitudeLoop)
list(PhaseLoop)
list(MagnitudeCL)
list(PhaseCL)

freq_Plant = np.array(MagnitudePlant['Frequency (Hz) - y0 (u0)'])
mag_Plant = np.array(MagnitudePlant['Magnitude (abs) - y0 (u0)'])
phase_Plant = np.array(PhasePlant['Phase (deg) - y0 (u0)'])



freq_Loop = np.array(MagnitudeLoop['Frequency (Hz) - y0 (u0)'])
mag_Loop = np.array(MagnitudeLoop['Magnitude (abs) - y0 (u0)'])
phase_Loop = np.array(PhaseLoop['Phase (deg) - y0 (u0)'])

freq_CL = np.array(MagnitudeCL['Frequency (Hz) - y0 (u0)'])
mag_CL = np.array(MagnitudeCL['Magnitude (abs) - y0 (u0)'])
phase_CL = np.array(PhaseCL['Phase (deg) - y0 (u0)'])



```

```python

freq_Loop = np.array(MagnitudeLoop['Frequency (Hz) - y0 (u0)'])
mag_Loop = np.array(MagnitudeLoop['Magnitude (abs) - y0 (u0)'])
phase_Loop = np.array(PhaseLoop['Phase (deg) - y0 (u0)'])

```

```python
for i in range(phase_Loop.size):
    phase_Loop[i] = phase_Loop[i] - 180
```

```python
phase_Loop
```

```python
fig_freq, (ax1, ax2) = plt.subplots(2, 1, sharex=True)
plt.legend()
fig_freq.suptitle('Frequency Response of Plant, Loop, and Closed-Loop', fontsize=16)
fig_freq.set_figheight(15)
fig_freq.set_figwidth(15)

ax1.semilogx(freq_Plant,20*np.log10(mag_Plant),'-b', label = 'Plant')
ax2.semilogx(freq_Plant,phase_Plant,'-b')

ax1.semilogx(freq_Loop,20*np.log10(mag_Loop),'-r', label = 'Loop Return Ratio')
ax2.semilogx(freq_Loop,phase_Loop,'-r')

ax1.semilogx(freq_CL,20*np.log10(mag_CL),'-g', label = 'Closed Loop')
ax2.semilogx(freq_CL,phase_CL,'-g')

ax1.set_yticks(np.arange(-40,60,20))
ax2.set_yticks(np.arange(-210,180,30))



ax1.set_xlabel('Frequency [Hz]')
ax1.set_ylabel('Magnitude [dB]')

ax2.set_xlabel('Frequency [Hz]')
ax2.set_ylabel('degrees °')
fig_freq.legend()
```

```python
#Plant

fig1, (ax1, ax2) = plt.subplots(2, 1, sharex=True)
fig1.suptitle('Frequency Response Plant', fontsize=16)
   

ax1.semilogx(freq_Plant,20*np.log10(mag_Plant),'.b')
ax2.semilogx(freq_Plant,phase_Plant,'.r')
ax2.set_yticks(np.arange(-210,30,30))


ax1.set_xlabel('Frequency [Hz]')
ax1.set_ylabel('Magnitude [dB]')

ax2.set_xlabel('Frequency [Hz]')
ax2.set_ylabel('degrees °')

```

```python


#Loop

fig2, (ax1, ax2) = plt.subplots(2, 1, sharex=True)
fig2.suptitle('Frequency Response Loop', fontsize=16)
   

ax1.semilogx(freq_Loop,20*np.log10(mag_Loop),'.b', label = 'data points')
ax1.semilogx(freq_Loop,np.ones(mag_Loop.size),'r--', label = 'cross-over frequency line')
ax1.legend(loc = 'upper right')

ax2.semilogx(freq_Loop,phase_Loop,'.r')



ax1.set_xlabel('Frequency [Hz]')
ax1.set_ylabel('Magnitude [dB]')

ax2.set_xlabel('Frequency [Hz]')
ax2.set_ylabel('degrees °')


plt.show()


# step response

```
